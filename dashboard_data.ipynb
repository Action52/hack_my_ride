```python
# from hmr_utils import *
import pandas as pd
import os
%matplotlib inline
```

```python
data_folder = './data/final_tagged_with_rm_forest_final/'
routes = []
for _file in os.listdir(data_folder):
    print(_file)
    route = pd.read_parquet(f"{data_folder}{_file}", engine='fastparquet')
    #route.dropna(subset=['qas_indi_real', 'qas_indi_pred'])
    routes.append(route)
routes = pd.concat(routes)
routes['period'] = routes['period'].astype(str)
routes
```

```python
routes['date'] = routes.index
routes.insert(0, 'date', routes.pop('date'))
stops_df = pd.read_csv('./data/gtfs3Sept/stops.txt')
merged_df = pd.merge(routes[['route_id', 'trip_headsign', 'direction_id', 'stop_id', 'date', 'headway', 'qas_indi_real', 'period']],
                     stops_df[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']],
                     on='stop_id')
merged_df['route_id'] = merged_df['route_id'].astype('int')
```

```python
merged_df
```

```python
routes_df = pd.read_csv('./data/gtfs3Sept/routes.txt')
merged_df = pd.merge(merged_df, routes_df[['route_id', 'route_short_name', 'route_long_name']], on='route_id')
merged_df['date'] = pd.to_datetime(merged_df['date']).dt.date
merged_df['period'] = pd.to_datetime(merged_df['period']).dt.time
merged_df = merged_df.rename(columns={'qas_indi_real': 'qas_gtfs'})
merged_df
```

```python
grouped_data = merged_df.groupby(['date', 'route_id', 'direction_id', 'stop_id', 'period']).apply(lambda x: x.mode())
```

```python
grouped_data = grouped_data.dropna()
grouped_data
```

```python
grouped_data.to_csv("dashboard_data.csv")
```
